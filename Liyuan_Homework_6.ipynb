## Problem 1

The code below generates a random sample from a Poisson distribution. Run the cell to create sample X. 

````python
import numpy as np
from scipy import stats

np.random.seed(13)
X = stats.poisson(7).rvs(925)
````

Use bootstrapping to find a 95\% confidence interval for the sample mean. Report your result with two after-decimal digit accuracy and justify all the choices you made in your code. 

````python
np.random.seed(10)

# create m = 1000 random bootstrap samples
m = 1000
bootstrap_X_mean = []  # initialize the variances matrix

# find sample means of each bootstrap sample
for i in range(m):
    bs_X = np.random.choice(X,size = 925) # create one bootstrap sample by resampling the original sample with replacement 100 times
    bootstrap_X_mean.append(np.mean(bs_X))

# find the lower-bound and upper-bound of sample mean estimates
lb = round(np.percentile(bootstrap_X_mean, 2.5),2)
ub = round(np.percentile(bootstrap_X_mean, 97.5),2)

print("The 95% confidence interval for the bootstrap sample mean is [",lb,",",ub,"].")
````

__Result:__ 
<br> The 95% condifence interval for the random bootstrap sample mean is around $[6.5, 7.55]$. 

<br>__Sample Size $n$:__ 
<br>Bootstrap samples are taken with replacement from the original sample. They usually have the same size $n$ as the original sample. In the original poisson sample, the sample size is equal to 925. Therefore, $size=925$ would be a good choice for this question.

<br>__Repeated Times $m$:__ 
<br> A bootstrap distribution of the statistic "sample mean" could be obtained by repeatedly finding the sample mean of very many bootstrap samples. To be specific, when $m = 1000$, the corresponding interval is $[6.84, 7.2]$. When $m = 10000$, results are very similar $[6.84, 7.18]$, which does not change a lot. Therefore, the repeated times $m = 1000$ would be a good choice for this question.

## Problem 2

Write a function called "local_peak" that takes as its input a one-dimensional Pandas series and returns a list of the positions of all local "peaks". That is, your function should return a list of the position values of all numbers in the series have have smaller values on both sides of them. 

Example: For ```pd.Series([1,3,2,4,4,5,7,4,5])``` your function should return ```[1,6]```.
For ```pd.Series([1,2,3,3,3,2])``` your function should return ```[]``` (since there is no local peak in this series, but rather a "plateau".)

````python
import numpy as np
import pandas as pd
X = pd.Series([1,3,2,4,4,5,7,4,5])
len(X)
````

````python
import pandas as pd
import numpy as np

def local_peak(X):
    """Finds position of local peaks in Pandas Series"""
    
    n = len(X)                             # get the lenth of Pandas series X
    local_peaks = []                       # initiate a list to store local peaks index

    for i in range(1,n-1):                        # for loop: tranverse the series index from 1 to (n-2)
            if X[i] > X[i-1] and X[i] > X[i+1]:   # if element is greater than both sides, it's a local peak
                    local_peaks.append(i)         # append the local peak index to the list
                    
    return local_peaks
            
# check your answer: 
X = pd.Series([1,3,2,4,4,5,7,4,5])
local_peak(X) == [1,6]
````
